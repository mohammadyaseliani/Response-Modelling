```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
data=pd.read_csv("conversion_data.csv")
data.head(2)
```

```python
# Remove the rowa containing null values
data.dropna()
```

```python
# Create the independent and dependent varaibles
X=data[['age','new_user','total_pages_visited']]
y=data['converted']
```

```python
# Create the training and test sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
# Extract the test set from data
ind_list = list(y_test.index)
data_test=data.iloc[ind_list]
```

```python
# Train logit model
from sklearn.linear_model import LogisticRegression
logit=LogisticRegression(max_iter=6000)
model_lr=logit.fit(X_train, y_train)
```

```python
# Predict probabilities using the test set
y_hats = model_lr.predict_proba(X_test)
```

```python
# Craete a column for probabilities 
data_test['Probabilities']=y_hats[:,1]
```

```python
# Sort Porabilities in desecnding order
data_sort=data_test.sort_values(by=['Probabilities'], ascending=False)
data_sort
```

```python
# Save the output as an excel file for further use and decision-making in Excel
data_sort.to_excel("output.xlsx") 
```
